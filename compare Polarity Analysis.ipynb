```python
import pandas as pd
```

```python
#TikTok_ Combine polarity analysis
```

```python
# Load the dataset- a CSV file - with a 'Lemmatized_Comments' column
df1 = pd.read_csv(r'C:\Users\paris\OneDrive\thesis\collect data\Analysis\data processing\Cleaned_TikTok_comments3.csv')[['Lemmatized_Comments']]
```

```python
# Load the dataset- a CSV file - with a 'RoBERTa_polarity' column
df2 = pd.read_csv(r'C:\Users\paris\OneDrive\thesis\collect data\Analysis\results\TikTok\Roberta_Emotion_Analysis_tiktok26.csv')[['RoBERTa_polarity']]
```

```python
# Load the dataset- a CSV file - with a 'VADER_polarity' column
df3 = pd.read_csv(r'C:\Users\paris\OneDrive\thesis\collect data\Analysis\results\TikTok\VADER_Emotion_Analysis_TikTok26.csv')[['VADER_Polarity']]
```

```python
# Load the dataset- a CSV file - with a 'TextBlob_polarity' column
df4 = pd.read_csv(r'C:\Users\paris\OneDrive\thesis\collect data\Analysis\results\TikTok\TextBlob_polarity_Emotion_Analysis_TikTok26.csv')[['TextBlob_polarity']]
```

```python
# Combine all the DataFrames horizontally
combined_df = pd.concat([df1, df2, df3, df4], axis=1)

# Display the combined DataFrame
print(combined_df.head())

# Save the combined DataFrame to a new CSV file
combined_df.to_csv(r'C:\Users\paris\OneDrive\thesis\collect data\Analysis\results\TikTok\combined_polarity_analysis.csv', index=False)
print("Combined DataFrame saved as 'combined_polarity_analysis.csv'.")
```

```python
# Combine polarity results into a single DataFrame
combined_df = combined_df[['Lemmatized_Comments', 'TextBlob_polarity', 'VADER_Polarity', 'RoBERTa_polarity']]

# Display the combined DataFrame for verification
print(combined_df.head())
```

```python
#Aggregate Polarity Counts, To understand the distribution of positive, negative, and neutral classifications for each method

# Calculate counts of each polarity for each method
polarity_counts = {
    'TextBlob': combined_df['TextBlob_polarity'].value_counts(),
    'VADER': combined_df['VADER_Polarity'].value_counts(),
    'RoBERTa': combined_df['RoBERTa_polarity'].value_counts()
}

# Convert to DataFrame for visualization
polarity_counts_df = pd.DataFrame(polarity_counts).fillna(0).astype(int)
print(polarity_counts_df)
```

```python
import matplotlib.pyplot as plt
# Plot bar chart to compare polarity distributions
polarity_counts_df.plot(kind='bar', figsize=(10, 6), color=['pink', 'orange', 'red'])
plt.title('Polarity Analysis Across Models on TikTok', fontsize=16)
plt.xlabel('Polarity', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=0, fontsize=14)
plt.legend(title='Model', fontsize=14)
plt.tight_layout()

# Save the plot
plt.savefig('polarity_comparison_bar_chart_TikTok.png')
plt.show()
```

```python

# Normalize the counts by calculating the percentage
polarity_counts_normalized = polarity_counts_df.div(polarity_counts_df.sum(axis=0), axis=1) * 100
print("Normalized Polarity Counts (in percentage):")
print(polarity_counts_normalized)

# Plot normalized polarity distributions
plt.figure(figsize=(10, 6))
polarity_counts_normalized.plot(kind='bar', figsize=(10, 6), color=['pink', 'orange', 'red'])
plt.title('Normalized Polarity Analysis Across Models on TikTok', fontsize=16)
plt.xlabel('Polarity', fontsize=14)
plt.ylabel('Percentage (%)', fontsize=14)
plt.xticks(rotation=0, fontsize=14)
plt.legend(title='Model', fontsize=14)
plt.tight_layout()

# Save the plot
plt.savefig('normalized_polarity_comparison_bar_chart_TikTok.png')
plt.show()
```
